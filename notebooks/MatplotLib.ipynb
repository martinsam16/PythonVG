```python
import matplotlib.pyplot as plt
```

# Gráfico de Linea

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

# Crear el grafico
plt.plot(x, y)

# Agregar etiquetas
plt.xlabel("Distancia (m)")
plt.ylabel("Tiempo (h)")
plt.title("Grafico de LLegada a VG")

# Mostrar el gráfico
plt.show()
```

# Gráfico de Dispersión

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

# Crear el grafico
plt.scatter(x, y)

# Agregar etiquetas
plt.xlabel("Distancia (m)")
plt.ylabel("Tiempo (h)")
plt.title("Grafico de LLegada a VG")

# Mostrar el gráfico
plt.show()
```

# Gráfico de barras

```python
labels = ["FC Barcelona", "Real Madrid", "Bayer"]
values = [10, 21, 13,]

# Crear el grafico
plt.bar(labels, values)

# Agregar etiquetas
plt.xlabel("Equipo")
plt.ylabel("Cantidad de goles")
plt.title("Goles x Equipo")

# Mostrar el gráfico
plt.show()
```

# Gráfico de pastel

```python
labels = ["FC Barcelona", "Real Madrid", "Bayer"]
values = [10, 21, 13,]

# Crear el grafico
plt.pie(values, labels=labels)

# Agregar etiquetas
plt.xlabel("Equipo")
plt.ylabel("Cantidad de goles")
plt.title("Goles x Equipo")

# Mostrar el gráfico
plt.show()
```
